# Guided Project: Visualizing Earnings Based On College Majors

This is a guided project from Datquest, with the objective to practice using matplotlib and pyplot, quickly creating visualisations to explore data. The data set used is based on the job outcomes of students who graduated from college between 2010 and 2012. The original aggregated data was released by American Community Survey, cleaned by FiveThirtyEight and released on their Github repository

```python
import matplotlib.pyplot as plt 
import pandas as pd
% matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
# returning the first row as a formatted table
recent_grads.iloc[0]
```

### Visualisation and exploration of data using histograms

The following questions will be evaluated:

* What percent of majors are predominantly male? Predominantly female?
* What's the most common median salary range?


#### DIstribution of Sample_size

```python
# DataFrame.hist() permits the manipulation of bins 
ax1 = recent_grads['Sample_size'].hist(bins=10, range=(0,5000))
ax1.set_xlabel("Sample_size")
ax1.set_ylabel("Frequecncy")
```

#### Distribution of Median

```python
ax2 = recent_grads["Median"].plot(kind='hist').set_xlabel("Median")
```

* The most common median salary range is between 30000 - 40000
* This can be validated using the value_counts() method 

```python
# DataFrame.head() method used to display the top 5 rows
recent_grads["Median"].value_counts().head()
```

* The most frequent median salary is 35000 followed by 40000

```python
recent_grads["Median"].median()
```

```python
recent_grads["ShareWomen"].value_counts(bins=2).sort_index(ascending=False)
```

### Visualisation and exploration of data using a scatter matrix plot

```python
from pandas.plotting import scatter_matrix
# Scatter_matrix while passing in a DataFrame with 2 columns returns a 2 by 2 scatter matrix plot 
scatter_matrix(recent_grads[["Sample_size", "Median"]], figsize = (10,10))
```

```python
#  Adjusting fig size for visibility
scatter_matrix(recent_grads[["Sample_size","Median","Unemployment_rate"]], figsize=(12,12))
```

```python
scatter_matrix(recent_grads[["Women","Median"]], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[["ShareWomen","Median"]], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[["Full_time","Median"]], figsize=(10,10))
```

* The scatter matrix plots show that majors that are predominantly female earn less

```python
recent_grads.head(10).plot.bar(x="Major", y="ShareWomen", legend=False).set_ylabel("ShareWomen")
```

```python
recent_grads.tail(10).plot.bar(x="Major", y="ShareWomen", legend=False).set_ylabel("ShareWomen")
```

* The percentage of women taking the last 10 majors are much higher than those taking the first 10 majors, with Communication Disorders Sciences and Services, and Early Childhood Education having greater than 90% of women
* The results show that engineering related majors have less than 40% of women

```python
recent_grads.head(10).plot.bar(x="Major", y="Unemployment_rate", legend=False).set_ylabel("Unemployment_rate")
```

```python
recent_grads.tail(10).plot.bar(x="Major", y="Unemployment_rate", legend=False).set_ylabel("Unemployment_rate")
```

* The results show that on average, the unemployment rate for the first 10 majors is less than the last 10 majors
* However, students who took the Nuclear Engineering major have the highest unemployment rate
